```
# -----------
# 1ª QUESTÃO:
# -----------

# Escreva um programa que seja capaz de identificar se uma palavra é palíndromo ou não. 
# Uma palavra palíndromo é a palavra que é possível se ler da esquerda para a direita e da direita para esquerda e tem o mesmo significado.
# Exemplo: ovo, reviver.

palavra = input("Entre com uma palavra: ").lower()
tamanho = len(palavra)

i = 0
j = tamanho - 1

while i < tamanho:
  if(palavra[i] == palavra[j]):
    palindromo = True
  else:
    palindromo = False
  i += 1
  j -= 1

if(palindromo):
  print("A palavra informada: '{}' é palídromo.".format(palavra))
else:
  print("A palavra informada: '{}' NÃO é palídromo.".format(palavra))


```

```
# -----------
# 3ª QUESTÃO
# -----------

# Uma classe de anagramas é um conjunto de palavras com as mesmas letras, mas por outra ordem.
# Exemplo: caras, casar e sacar são anagramas, fio e foi formam outra classe.
# Escreva um programa que exiba todas as classes de anagramas e suas palavras a partir da leitura de um arquivo texto.

import itertools

caracteresEspeciais = "!@#$%¨&*()_+-={}][`^´~<>:,.;|\/?°ºª'"

texto = "Caras! Hoje acordei com uma sensação de cansado por dormir tarde em razão dos estudos, mas foi muito bom!".lower()

for caracter in caracteresEspeciais:
  texto = texto.replace(caracter, "")

palavras = texto.rsplit(" ")

vetorDeLetras = []
classeDeAnagramas = []

for palavra in range(len(palavras)):

  i = 0
  while i < len(palavras[palavra]):
    vetorDeLetras.append(palavras[palavra][i])
    i += 1

  listaDePalavras = list(itertools.permutations(vetorDeLetras))
  vetorDeLetras.clear()
  
  j = 0
  while j < len(listaDePalavras):
    
    anagrama = ""
    for letra in listaDePalavras[j]:
      anagrama += letra
    
    classeDeAnagramas.append(anagrama)
    j += 1
  
  print("Classe de anagramas: ", classeDeAnagramas)
  classeDeAnagramas.clear()
  
```
